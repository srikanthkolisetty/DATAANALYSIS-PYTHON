# Preprocessing 

## Robust Scaling
* If our dataset contains more number of outliers then robust scaling is the best technique to remove the outliers.

```python
# xi--Q2(x)
#-------------
# Q3(x)-Q1(x)
#Q1-->25% value
# Q2-->median
# Q3--->75% value
```

```python
import pandas as pd
import numpy as np
```

```python
file=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
file.head()
```

```python
file.drop(['Unnamed: 0'],inplace=True,axis=1)
```

```python
file
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs=RobustScaler()
rs
```

```python
rs2=rs.fit_transform(file)  # applys the robust scalar formula to the data.


rs2
```

```python
rs2=pd.DataFrame(rs2,columns=file.columns)
rs2.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig,(axis1,axis2) =plt.subplots(ncols=2,figsize=(9,5))
sns.kdeplot(file['TV'],ax=axis1)
sns.kdeplot(file['radio'],ax=axis1)
sns.kdeplot(file['newspaper'],ax=axis1)
sns.kdeplot(file['sales'],ax=axis1)

sns.kdeplot(rs2['TV'],ax=axis2)
sns.kdeplot(rs2['radio'],ax=axis2)
sns.kdeplot(rs2['newspaper'],ax=axis2)
sns.kdeplot(rs2['sales'],ax=axis2)
```

```python
file.quantile(0.25)
```

### Standard Scaler
* Most commonly used scaling  tehnique.It  scales between 0 to  mean value

```python
# xi-mean(x)
# ----------------
# std(x)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
```

```python
s2=s.fit_transform(file)
```

```python
s2=pd.DataFrame(s2,columns=file.columns)
s2
```

```python
fig,(axis1,axis2) =plt.subplots(ncols=2,figsize=(9,5))
sns.kdeplot(file['TV'],ax=axis1)
sns.kdeplot(file['radio'],ax=axis1)
sns.kdeplot(file['newspaper'],ax=axis1)
sns.kdeplot(file['sales'],ax=axis1)

sns.kdeplot(s2['TV'],ax=axis2)
sns.kdeplot(s2['radio'],ax=axis2)
sns.kdeplot(s2['newspaper'],ax=axis2)
sns.kdeplot(s2['sales'],ax=axis2)
```
